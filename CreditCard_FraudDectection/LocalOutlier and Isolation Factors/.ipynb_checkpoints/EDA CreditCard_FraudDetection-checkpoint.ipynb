# Credit Card Fraud Detection
### Dataset is hosted in Kaggle 
This algotrithm sample will try to detect fraudulent credit card transactions.  In this Fintech project, we will be using a dataset of of nearly 28,500 credit card transactions (subset of real  life credit card transactions) hosted in Kaggle.com. In the interest of Security, the transactions are hashed to protect the customer data. In this project, we will build and deploy the following two machine learning algorithms:
- Local Outlier Factor (LOF)
- Isolation Forest Algorithm

Furthermore, using metrics suchs as precision, recall, and F1-scores, we will investigate why the classification accuracy for these algorithms can be misleading.

In addition, we will explore the use of data visualization techniques common in data science, such as parameter histograms and correlation matrices, to gain a better understanding of the underlying distribution of data in our data set. Let's get started

### 2. The Data Set

In the following cells, we will import our dataset from a .csv file as a Pandas DataFrame.  Furthermore, we will begin exploring the dataset to gain an understanding of the type, quantity, and distribution of data in our dataset.
```python
# Load the dataset from the csv file using pandas
dataset = pd.read_csv('creditcard.csv')
```

```python
# Start exploring the dataset
print(dataset.columns)
```

### REDUCE THE DATASET
In the interest of resource reduction, we will be reducing the dataset to 10% for faster processing 

```python
# V1 - V28 are the results of a PCA Dimensionality reduction to protect user identities and sensitive features
data_frac = dataset.sample(frac=0.1, random_state = 1)
print(data_frac.shape)
data_frac.describe()
```

```python
sns.pairplot(data_frac, hue = 'Class', vars = ['V1', 'V2', 'V3', 'V4', 'V5'])
```

```python
plt.figure(figsize=(10,10)) 
sns.heatmap(dataset.corr(), annot=False) 
```

```python
# Plot histograms of each parameter 
data_frac.hist(figsize = (20, 20))
plt.show()
```
